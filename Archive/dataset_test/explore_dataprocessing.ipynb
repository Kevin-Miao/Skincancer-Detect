## Understanding the data

Our data come from 4 different distinct datasets.
- Vidir Modern
- RosenDahl
- Vienna Dias
- Vidir Molemax

If interested, we can always augment our dataset accordingly by adding in pictures from other dataset, but it is good to start at this baseline.

```python
metadata['dataset'].unique()
```

The following diseases are being diagnoses:

```python
sorted(metadata['dx'].unique())
```

**Legend**:
- `akiec`: Actinic Keratosis 

        This disease is common and non-invasive. They can be treated without locally without surgery. Only small fraction becomes malignant.
- `bcc`: Basal Cell Carcinoma

        This disease is a common variant of ephitelial skin cancer that rarely metastasizes or grow destructively if untreated.
- `bkl`: Benign Keratosis   

        Benign supergroup of different kinds of warts

- `df`: Dermatofibroma

        Benign skin lesion regardes as either a benign proliferation or an inflammatory reaction to minimal trauma.

- `nv`: Melanocytic nevi

        Benign neoplasms of melanocytes. If diagnosed properly, are biologically stable, completely benign lesions. However, melanocytic nevi can be found in association with melanoma

- `mel`: Melanoma

        **Malignant** neoplasms derived from melanocytes. In early stage it is treatable, but can become highly invasive in the future.

- `vasc`: Vascular Skin Lesions
        
        They can be both malignant or benign.
